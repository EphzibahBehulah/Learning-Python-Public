This is what we will be doing here. We will connect to our database in Bigquery, and then run queries from our google colab.

```python
# 1️⃣ Select the highest overall ranking player
query1 = '''
SELECT overall
FROM `redi-da25.fifa.players_22`
ORDER BY overall DESC
LIMIT 1;
'''
df.head(1)
```

```python
# 2️⃣ Select the player who earns the most
query2 = '''
SELECT wage_eur
FROM `redi-da25.fifa.players_22`
ORDER BY wage_eur DESC
LIMIT 1;
'''
df.head(1)

```

```python
# 3️⃣Select players who have 'Lionel' in their long name
query3 = '''
SELECT *
FROM `redi-da25.fifa.players_22`
WHERE long_name LIKE '%Lionel%';
'''
df.head(1)
```

```python
# 4️⃣ Count how many players have 'Lionel' in their long name
query4 = '''
SELECT COUNT(*) AS num_players_with_lionel
FROM `redi-da25.fifa.players_22`
WHERE long_name LIKE '%Lionel%';
'''
df.head(1)

```

```python
# 5️⃣ Select the 10 heaviest players with position 'RW' (Right Wing)
query5 = '''
SELECT *
FROM `redi-da25.fifa.players_22`
WHERE position = 'RW'
ORDER BY weight_kg DESC
LIMIT 10;
'''
df.head(5)
```
